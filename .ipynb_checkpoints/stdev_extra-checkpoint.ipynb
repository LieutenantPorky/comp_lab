## Part 1 - modelling poulation growth

### Theory
Population dynamics are a dynamical system, whose change depends on the current size of the population as well as certain constants describing the system. A simple iterative model can be constructed by considering the growth of the next generation as being directly proportional to the current population number by some scalar multiple, indicating the relative population growth (the ratio of births to deaths given infinite resources). However, it is fairly evident that such a population would continue growing exponentially. Real-world systems have an additional constraint imposed by finite resource availability, which slows down population growth as the population size approaches a limit (the carrying capacity). One can describe the next iteration of this population model as

$$ x_{n+1} = a \times x_{n} \times (1 - x_{n}) $$

where $x_{n}$ and $x_{n + 1}$ are, respectively, the size of the current and next generation as a fraction of the carrying capacity, and $a$ is the ratio of the birth and death rate under limitless resources.

One would expect that such a model would eventually reach an equilibrium, dictated by the carrying capacity - indeed, this is the case if one cosiders the changes to be derivatives, and treats the model as a differential equation. However, for finite-difference models, this does not always occur, and the population may instead continuously oscillate or behave chaotically after certain critical values of the expansion coefficient $a$.

### Implementation
To facilitate future modification, the algorithm has been designed with modularity in mind. The long-term equilibrium of a given population is calculated by `getConvergence()`, a functional returning the result of the repeated application of some input function to an initial value to determine convergence. The input function can be separately defined, to allow easy modification of test models, but is in this case crated by `getPopFunction()`, which returns the stated population function for a given value of $a$. `getConvergence()` is used in a loop (through python list interpretation) inside `randomCurve()` to generate a theoretical convergence value for a random initial condition and a given value of $a$. It should be noted that the function need not converge, in which case the final value will be pseudorandom, determined chaotically by the initial value and the iteration number. This process is repeated through another for loop to obtain 50 datapoints at a certain value of $a$, which is swept over the required range. The resultant convergence values are then stored in `aSweep`, a matrix of ($a$ value, convergence) pairs:

$$ 
\mathrm{aSweep} = 
\begin{bmatrix}
[a_{0}, \mathrm{conv.}_{0}] & [a_{0}, \mathrm{conv.}_{1}] & [a_{0}, \mathrm{conv.}_{2}]  & \dots\\
[a_{1}, \mathrm{conv.}_{0}] & [a_{1}, \mathrm{conv.}_{1}] & [a_{1}, \mathrm{conv.}_{2}]  & \dots\\
[a_{2}, \mathrm{conv.}_{0}] & [a_{2}, \mathrm{conv.}_{1}] & [a_{2}, \mathrm{conv.}_{2}]  & \dots\\
\vdots & \vdots & \vdots & \ddots \\
\end{bmatrix}
$$

These values are then unpacked into two vectors of index-wise $a$ and convergence values, and plotted as single points using matplotlib.

```python
#This function returns the population growth function for a given "a" value
def getPopFunction(a):
    return lambda n : a * n * (1 - n)

#Determine the convergence of a certain population function after an arbitrary number of iterations:

def getConvergence(initial, popFunction, iterations):
    for i in range(0,iterations):
        initial = popFunction(initial)
    return initial

#Initialize a population with a random size, and iterate with a = 1 to find the convergence (with 100 iterations)

def randomCurve(a, iterations):
    return [a, getConvergence(np.random.random(),getPopFunction(a), iterations)]


#now, iterate that function through 2000 values of a in [1.5,4]
aSweep = [[randomCurve(a,100) for i in range(0,50)] for a in np.linspace(1.5,4,2000)]

#unpack and plot the obtained values, reducing the point size to 0.1 to resolve detail
ax = plt.subplot(111)
ax.plot([[j[0] for j in i] for i in aSweep], [[j[1] for j in i] for i in aSweep],'bo', ms=0.1)
plt.show()
```

```python
#Same as above, but for a in [3,4] and [3.5,3.8]
aSweep2 = [[randomCurve(a,100) for i in range(0,50)] for a in np.linspace(3,4,2000)]

ax1 = plt.subplot(211)
ax1.set_ylim(0,1)
ax1.plot([[j[0] for j in i] for i in aSweep2], [[j[1] for j in i] for i in aSweep2],'bo', ms=0.1)

aSweep2 = [[randomCurve(a,100) for i in range(0,50)] for a in np.linspace(3.5,3.8,2000)]

ax2 = plt.subplot(212)
ax2.set_ylim(0,1)
ax2.plot([[j[0] for j in i] for i in aSweep2], [[j[1] for j in i] for i in aSweep2],'bo', ms=0.1)

plt.show()
```

### Quantifying chaos



```python
#We need to divide the area into regions and calculate an average numerical density
aSize = 200
cSize = 50

aWidth = (4 - 1.5) / aSize
cWidth = 1 / cSize

aSweep = [[randomCurve(a,100) for i in range(0,50)] for a in np.linspace(1.5,4,1000)]

densityMatrix = [[0 for i in range(0,cSize)] for j in range(0,aSize)]
aVector = [] # transform values from a matrix to a vector
for i in aSweep:
    aVector = aVector + i

#for each cell, figure out how many  a, convergence pairs are enclosed

for aIndex, aValue in enumerate(np.linspace(1.5,4,aSize, endpoint=False)):
    for cIndex, cValue in enumerate(np.linspace(0,1,cSize, endpoint = False)):
        regionNum = 0        
        # check all a, conv pairs, and add them to the region counter
        
        for index, pair in enumerate(aVector):
            #print(pair)
            if (aValue + aWidth > pair[0] and aValue < pair[0]) and (cValue + cWidth > pair[1] and cValue < pair[1]): #if the point is inside the region
                regionNum += 1
                del aVector[index] # delete points we added, to increase efficiency on future loops
        densityMatrix[aIndex][cIndex] = regionNum # update the corresponding density

plt.plot(np.linspace(1.5,4,aSize, endpoint=False), [np.std(i) for i in densityMatrix])
plt.show()
```

```python
#We want to calculate the density of measured convergence values at different a's
regions = 50 # How many "areas" we use to estimate density variance
groupSize = 1.0 / regions # the size of each group

densityMatrix = [] # matrix to hold densities

#aSweep = [[randomCurve(a,100) for i in range(0,500)] for a in np.linspace(1.5,4,1000)]


for index, aVal in enumerate(aSweep):
    # We know convergence values will lie between 0 and 1, so we can break this interval into small regions:
    currentA = aVal[0][0]
    densityVector = [] # hold the group densities at this value of a
    for group in np.linspace(0,1,regions, endpoint=False):
        groupSize = 0
        # For efficiency, we want to eliminate any values we're already checking for each aVal array
        for i, convergenceValue in enumerate(aVal):
            #print(convergenceValue[1])
            if group + groupSize > convergenceValue[1]: # if the convergence value is in the group                
                groupSize += 1
                del aSweep[index][i] # remove item from list for better efficiency
        #Now, add the groupsize to the vector
        densityVector.append([currentA, groupSize])
    densityMatrix.append(densityVector)

density_std = [np.std([i[1] for i in row]) for row in densityMatrix]
a_vals = [row[0][0] for row in densityMatrix]

plt.plot(a_vals, density_std)
plt.show()
```

### Convergence and chaotic phenomena

As can be seen from the produced graphs, the population dynamics are fairly predicatable for low values of $a$, with any initial value converging to a single fixed point. However, this convergence does not occur when the growth constant $a$ surpasses a certain constant, with the long term behaviour of the population being a regular oscillatory system for intermediate values or chaotic for larger values. Interestingly, this effect is not observed if we consider a modification of the model based on differential changes, and then attempt to find a stable numerical solution to the obtained differential equation (using scipy's 4th-order RK solver), and the chaotic effects observed require a finite-difference model.

```python
import scipy.integrate

#Example of how oscillatory behaviour only occurs in finite-difference, euler integration - A model based on infinitesimal
#growth does not demonstrate long-term oscillatory patterns, and instead will converge.

diffSolution = scipy.integrate.solve_ivp(lambda t,y : 3.2 * y * (1 - y), [0,10], [0.5], t_eval = np.linspace(0,10,50))

plt.plot(diffSolution.t, diffSolution.y[0])
plt.show()
```

```python
#Plot specific populations at different values of a: some will converge, some will oscillate regularly,
#and some are chaotic

#Apply the population growth function every step, but this time save the population number every time, not just at the end

def getEvolution(initial, function, iterations):
    datapoints = [[0, initial]]
    for i in range(1,iterations):
        datapoints.append([i, function(datapoints[i-1][1])])
    return datapoints

ax0 = plt.subplot(411)
ax1 = plt.subplot(412)
ax2 = plt.subplot(413)
ax3 = plt.subplot(414)

#Plot populations for a= 3.8(chaotic), 3.5 (oscillatory on 4 points), 3.2 (oscillatory on 2 points) and 2.5 (converges cleanly)

path = getEvolution(0.5, getPopFunction(3.8), 100)
ax0.plot([i[0] for i in path], [i[1] for i in path])
ax0.set_title(r"$a$ = 3.8")

path = getEvolution(0.5, getPopFunction(3.5), 100)
ax1.plot([i[0] for i in path], [i[1] for i in path])
ax1.set_title(r"$a$ = 3.5")

path = getEvolution(0.5, getPopFunction(3.2), 100)
ax2.plot([i[0] for i in path], [i[1] for i in path])
ax2.set_title(r"$a$ = 3.2")

path = getEvolution(0.5, getPopFunction(2.5), 100)
ax3.plot([i[0] for i in path], [i[1] for i in path])
ax3.set_title(r"$a$ = 2.5")

plt.tight_layout()
plt.show()
```

### Part 2 - montecarlo integration

```python
plt.ion()
ax1 = plt.subplot(111)
def montecarloIntegrate(function, bounds):
    #First thing we need to do is find the bounding box of the function - we know that sin^4 is bounded by [0,1], but
    #arbitrarily choosing these values is not very generalizable (also it's cheating -_-)
    
    testSeq = [function(i) for i in np.linspace(bounds[0], bounds[1], 200)]
    plt.plot(np.linspace(bounds[0], bounds[1], 200), testSeq)
    
    #find the global maxima and minima of the function
    gMax = np.amax(testSeq)
    gMin = np.amin(testSeq)
    diff = gMax - gMin

    # now we have the vertical bounds - we just need to create a bunch of random points in the area and check whether
    #they are smaller than the function

    iterations = 10000
    numInside = 0.0
    
    
    for x,y in zip(np.random.random(iterations) * (bounds[1] - bounds[0]) + bounds[0], np.random.random(iterations) * diff + gMin):
        value = function(x)            
        #Firstly, the point must have the same sign as the function to be under the curve:
        if (value * y >= 0):
            # if it has the same sign, check whether it is under the curve
            if (np.abs(value) > np.abs(y)):
                #Now, if the value is negative, subtract a point from the total. Else, add a point
                if value < 0:
                    plt.plot(x,y,'bo')
                    numInside -= 1
                else:
                    numInside += 1
                    plt.plot(x,y,'ro')
            else:
                plt.plot(x,y,'ko')
    
    # Now, find ratio of points inside and outside the bounds, and multiply by total area:
    
    return numInside / iterations * diff * (bounds[1] - bounds[0])

print(montecarloIntegrate(lambda x : np.sin(x) ** 4, [0.0, 0.5 * np.pi]))
plt.show()
```

### Manually computing the integral

We consider the integral
$$\int \sin^{4}x \mathop{}\!\mathrm{d}x$$

The sine can be split, and then integration by parts can be applied

$$\int \sin^{4}x \mathop{}\!\mathrm{d}x = \int \sin^{3}x \sin x \mathop{}\!\mathrm{d}x = -\sin^{3}x \cos x + \int 3\sin^{2}x \cos^{2} x \mathop{}\!\mathrm{d}x $$

Noting that $ 2\sin x \cos x = \sin 2x$,

$$ -\sin^{3}x \cos x + \int \frac{3}{4}\sin^{2}2x \mathop{}\!\mathrm{d}x$$

$\cos 2x = \cos^{2} x - \sin^{2} x = 1 - 2\sin^{2}x$, allowing us to express $\sin^{2}$ in terms of $\cos 2x$. This then allows us to finally compute the integral.

$$-\sin^{3}x \cos x + \int \frac{3}{8}(1 - \cos 4x) \mathop{}\!\mathrm{d}x =
-\sin^{3}x \cos x + \frac{3}{8}(x - \frac{\sin 4x}{4}) + C$$

Collecting the terms,

$$-\sin^{3}x \cos x + \frac{3}{8}x - \frac{3\sin 4x}{32} + C$$

We can simplify this expression:

$$ \sin^{3}x \cos x = \sin^{2} x \sin x \cos x = \frac{1}{2}(1 - \cos 2x)\frac{1}{2}(\sin 2x) = \frac{1}{4}\sin 2x - \frac{1}{8} \sin 4x$$

Plugging back into our original expression,

$$-(\frac{1}{4}\sin 2x - \frac{1}{8} \sin 4x) + \frac{3}{8}x - \frac{3\sin 4x}{32} =
\frac{3}{8}x - \frac{1}{4}\sin 2x + \frac{1}{32}\sin 4x$$

Therefore,

$$ \int \sin^{4}x \mathop{}\!\mathrm{d}x = \frac{3}{8}x - \frac{1}{4}\sin 2x + \frac{1}{32}\sin 4x + C $$

For the specific example, we are considering the definite integral

$$\int^{\pi / 2}_{0} \sin^{4}x \mathop{}\!\mathrm{d}x$$

which can be computed through the above indefinite integral to be $\frac{3\pi}{16} \approx 0.589$

### Montecarlo in 3D

```python

#For 3 dimensions, the integration path/volume/surface is a function
def isShell(point):
    r1 = 0.75
    r2 = 1
    return sum([i ** 2 for i in point]) >= r1 ** 2 and sum([i ** 2 for i in point]) <= r2 ** 2

def integrateVolume(volume, samples):
    total = 0.0
    for i in range(0,samples):
        #create a random point in 3d space, with coordinates in [0,1]
        point = [2 * i - 1 for i in np.random.random(2)]
        if volume(point):
            total += 1
    return total / samples * 2**3

print (integrateVolume(isShell, 10000))
```
